```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
from ivis import Ivis
```

```python
x, y = make_blobs(n_samples=1000, n_features=2)
plt.scatter(x.T[0], x.T[1], c=y);
```

```python
x, y = make_blobs(n_samples=1000, n_features=3)
ivis = Ivis(embedding_dims=2, k=5)  # k = 0.5-1% of data
```

```python
x.shape
```

```python
x = MinMaxScaler().fit_transform(x)
x = ivis.fit_transform(x)
plt.scatter(x.T[0], x.T[1], c=y);
```

We now have 3 separable clusters. Let's check this with a high dimensional dataset, say 10 dimensions.

```python
x, y = make_blobs(n_samples=1000, n_features=10)
ivis = Ivis(embedding_dims=2, k=5)  # k = 0.5-1% of data

x = MinMaxScaler().fit_transform(x)
x = ivis.fit_transform(x)
plt.scatter(x.T[0], x.T[1], c=y);
```

It works well! Let's now use the DBSCAN trick.

```python
from sklearn.cluster import DBSCAN
import numpy as np
```

```python
cluster = DBSCAN(eps=0.07, min_samples=3)
cluster.fit(x)

xb = x[np.where(cluster.labels_ == -1)[0]]
yb = y[np.where(cluster.labels_ == -1)[0]]
print(len(xb))
plt.scatter(xb.T[0], xb.T[1], c=yb)
```
