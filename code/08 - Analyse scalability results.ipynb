```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../experiment_results/final-ant/scalability.txt', sep=' ', header=None)
```

```python
df.head()
```

```python
df['size'] = [x.split('_')[1] for x in df[0]]
```

```python
df['model'] = [x.split('_')[0] for x in df[0]]
```

```python
df.drop(0, axis=1, inplace=True)
```

```python
df.head()
```

```python
[x for x in df.iterrows()]
```

```python
df['p25'] = [np.percentile(row[:20], 25) for _, row in df.iterrows()]
```

```python
df['p50'] = [np.percentile(row[:20], 50) for _, row in df.iterrows()]
```

```python
df['p75'] = [np.percentile(row[:20], 75) for _, row in df.iterrows()]
```

```python
df.head()
```

```python
df.drop(range(1, 21), axis=1, inplace=True)
```

```python
df.head()
```

```python
groups = df.groupby('model')
```

```python
groups = list(groups)
```

```python
groups[:2]
```

```python
plt.figure(figsize=(12, 10))
plt.subplots_adjust(hspace=0.5, wspace=0.3)

for i, (model, data) in enumerate(groups):
    plt.subplot(3, 3, i + 1)
    plt.errorbar(data['size'], data['p50'], yerr=(data['p50'] - data['p25'], data['p75'] - data['p50']), 
                 ecolor='red', marker='.')
    plt.ylabel('Runtime')
    plt.xlabel('Size')
    plt.title(model)
```
